### Import all required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Load iris datasets
iris = load_iris()
X = pd.DataFrame(iris.data, columns = iris.feature_names)
X.head(10)
```

```python
X.info()
```

```python
X.describe()
```

```python
X.isnull().sum()
```

```python
y = iris.target
y
```

## Split Dataset into train and test sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Defining the Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)

print('Decision Tree Classifier Created successfully')
```

```python
y_pred = dt.predict(X_test)
```

```python
y_pred
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn import tree
```

## visualizing the Decision Tree

```python
fn = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
cn = ['setosa', 'versicolour', 'verginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 350)

tree.plot_tree(dt, feature_names = fn, class_names = cn, filled = True);
```
